## Creating a basemap of Washington D.C. to use for the analysis

initializing basic world map

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap()
map.drawcoastlines()

plt.show()
```

default map uses Plate Carree proejction centered at (lon., lat.)= (0,0)

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua
plt.show()
```

Same map, but now lakes and oceans are "aqua" and land is "coral"

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(lat_0= 38.8951100, lon_0= -77.0363700)
map.drawcoastlines()
map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua

x, y= map(-77.0363700, 38.8951100)
map.plot(x, y, marker="o", color='m')

plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
map = Basemap(resolution= 'f',projection= 'aea',llcrnrlon=-77.12059, llcrnrlat= 38.788345, urcrnrlon=-76.904984, urcrnrlat=38.995173,lat_0= 38.8951100, lon_0= -77.0363700)
map.drawcoastlines()
map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua

map.readshapefile('D:\OneDrive - University of South Carolina - Moore School of Business\DC_crime\Data\Ward_from_2012\Ward_from_2012', 'ward_shapefile')

x, y= map(-77.0363700, 38.8951100)
map.plot(x, y, marker="o", color='m')

plt.show()
```

using westlimit=-77.12059; southlimit=38.788345; eastlimit=-76.904984; northlimit=38.995173 to display only Washington DC. this information was accessed from http://boundingbox.klokantech.com/
