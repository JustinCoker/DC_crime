```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
map = Basemap(resolution= 'f',projection= 'aea',llcrnrlon=-77.12059, llcrnrlat= 38.788345, urcrnrlon=-76.904984, urcrnrlat=38.995173,lat_0= 38.8951100, lon_0= -77.0363700)
map.drawcoastlines()
map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua

map.readshapefile('D:\OneDrive - University of South Carolina - Moore School of Business\DC_crime\Data\Ward_from_2012\Ward_from_2012', 'ward_shapefile', drawbounds = False)

x, y= map(-77.0363700, 38.8951100)
map.plot(x, y, marker="o", color='m')

for info, shape in zip(map.ward_shapefile_info, map.ward_shapefile):
    if info['WARD'] == 1:
        x, y = zip(*shape) 
        map.plot(x, y, marker=None,color='m')

plt.show()
```
