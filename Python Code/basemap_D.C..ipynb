## Creating a basemap of Washington D.C. to use for the analysis

initializing basic world map

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap()
map.drawcoastlines()

plt.show()
```

default map uses Plate Carree proejction centered at (lon., lat.)= (0,0)

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua
plt.show()
```

Same map, but now lakes and oceans are "aqua" and land is "coral"

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(lat_0= 38.8951100, lon_0= -77.0363700)
map.drawcoastlines()
map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua

x, y= map(-77.0363700, 38.8951100)
map.plot(x, y, marker="o", color='m')

plt.show()
```

same map, centered on washington DC (purple circle) lat_1=45.,lat_2=55

```python
#Importing libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(lat_1=45.,lat_2=55,lat_0= 38.8951100, lon_0= -77.0363700)
map.drawcoastlines()
map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua

x, y= map(-77.0363700, 38.8951100)
map.plot(x, y, marker="o", color='m')

plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
map = Basemap(resolution= 'f',projection= 'aea',width=1200000,height=900000,lat_1=45.,lat_2=55,lat_0= 38.8951100, lon_0= -77.0363700)
map.drawcoastlines()
map.drawmapboundary(fill_color= 'aqua') #fills map with aqua
map.fillcontinents(color='coral', lake_color='aqua') #fills continents with coral and lakes with aqua

x, y= map(-77.0363700, 38.8951100)
map.plot(x, y, marker="o", color='m')

plt.show()
```
